```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
caminho = 'data\\destinations.csv'
df_tour = pd.read_csv(caminho, sep = ',', encoding = 'ISO-8859-1')

```

```python

df_tour = df_tour.dropna(subset=['Latitude'])

# Ordenar pelo valor da latitude (maior = mais ao norte)
df_norte = df_tour.sort_values(by='Latitude', ascending=False).head(8)
```

```python
plt.figure(figsize=(10,6))
plt.barh(df_norte['Destination'], df_norte['Latitude'], color='skyblue')
plt.xlabel('Latitude (graus)')
plt.ylabel('Destino')
plt.title('Os 8 destinos tur√≠sticos mais ao norte da Europa')
plt.gca().invert_yaxis()  # inverter para o mais ao norte aparecer no topo
plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
