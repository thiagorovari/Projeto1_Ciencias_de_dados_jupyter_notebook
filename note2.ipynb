```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/destinations.csv', encoding='ISO-8859-1')
# Normalização mínima
for col in df.columns:
    if df[col].dtype == object:
        df[col] = df[col].str.strip()
```

```python
col_country = None
col_destination = None
candidates_country = ['Country','country','COUNTRY']
candidates_destination = ['Destination','destination','Name','name']
for c in candidates_country:
    if c in df.columns:
        col_country = c
    break
for c in candidates_destination:
    if c in df.columns:
        col_destination = c
    break

df_unique = df.drop_duplicates(subset=['Country', 'Destination'])
destinos_por_pais = df_unique.groupby('Country')['Destination'].nunique().sort_values(ascending=False).head(15)
```

```python
plt.figure(figsize=(10,6))
destinos_por_pais.plot(kind='bar')
plt.title('Top 15 países por número de destinos (únicos)')
plt.ylabel('Número de destinos únicos')
plt.xticks(rotation=45)
plt.show()
```
